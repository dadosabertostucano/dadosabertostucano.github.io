```python
#Importando biblioteca para trabalhar
import pandas as pd
import numpy as np 
df_diaria = pd.read_csv('diarias-2020.csv', encoding='utf-8')
df_diaria.head()
```

```python
#Eliminando colunas indesejadas 
df_diaria = df_diaria.drop(['ï»¿""', 'SqLiquidacao', 'SqDiaria', 'ParentForm', 'TpOrigem' ], axis=1)
df_diaria.head()
```

```python
df_diaria['Diária'] = df_diaria['Diária'].str.replace(',', '.').astype(float)
df_diaria['Total'] = df_diaria['Total'].str.replace(',', '.').astype(float)
df_diaria.info()
```
