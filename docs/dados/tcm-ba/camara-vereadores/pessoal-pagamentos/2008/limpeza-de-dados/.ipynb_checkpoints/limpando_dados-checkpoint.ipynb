```python
#importe a biblioteca usada para consultar uma URL
import urllib.request

#importe as funções BeautifulSoup para analisar os dados retornados do site
from bs4 import BeautifulSoup
```

```python
#C:\Users\Andre\Desktop\Dados2
mes = 'jan'
#C:\Users\Aluno\Dropbox\Dados\Dados2\Novo_IRA
wiki_2 = 'file:///C:/Users/Micro/Documents/repo/dadosabertostucano.github.io/docs/\
dados/tcm-ba/camara-vereadores/pessoal-pagamentos/2008/limpar_html/jan-2008.html'
page_2 = urllib.request.urlopen(wiki_2)

soup = BeautifulSoup(page_2, 'html5lib')
all_table = soup.find_all('table')
#table = soup.find('table')
#table = soup.find('table', attrs={'class': 'table table-striped'})
#class="listagem"

tabelas = []
for tabela in soup.find_all('table', attrs={'class': 'table table-striped'}):
    tabelas.append(tabela) 

table = tabelas[1]   
    
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]


for row in table.findAll("tr"): #para tudo que estiver em <tr>
    cells = row.findAll('td') #variável para encontrar <td>
    if len(cells)==8: #número de colunas
        A.append(cells[0].find(text=True)) #iterando sobre cada linha
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
        G.append(cells[6].find(text=True))
        H.append(cells[7].find(text=True))

import pandas as pd

df = pd.DataFrame(columns=['Sinal'])

df['Sinal']=A
df['Nome']=B
df['Matrícula']=C
df['Tipo Servidor']=D
df['Cargo']=E
df['Salário Base']=F
df['Salário Vantagens']=G
df['Salário Gratificação']=H


df.head()
```

```python
#Eliminando colunas indesejadas 
df = df.drop(['Sinal'], axis=1)
df.head()
```

```python
#Inserindo valores para matrícula 
df['Matrícula'] = '000'
df['Salário Vantagens'] = 'R$ 000,00'
df['Salário Gratificação'] = 'R$ 000,00'
df.head()
```

```python
df['Salário Base'] = df['Salário Base'].apply(lambda x: x.split('R$')[1])
df['Salário Vantagens'] = df['Salário Vantagens'].apply(lambda x: x.split('R$')[1])
df['Salário Gratificação'] = df['Salário Gratificação'].apply(lambda x: x.split('R$')[1])

df['Salário Base'] = df['Salário Base'].apply(lambda x: x.split()[0].replace('.', ''))
df['Salário Vantagens'] = df['Salário Vantagens'].apply(lambda x: x.split()[0].replace('.', ''))
df['Salário Gratificação'] = df['Salário Gratificação'].apply(lambda x: x.split()[0].replace('.', ''))

df['Salário Base'] = df['Salário Base'].str.replace(',', '.').astype(float)
df['Salário Vantagens'] = df['Salário Vantagens'].str.replace(',', '.').astype(float)
df['Salário Gratificação'] = df['Salário Gratificação'].str.replace(',', '.').astype(float)

df.head()
```
